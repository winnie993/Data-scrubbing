```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
myurl='https://jiji.co.ke/nairobi/cars?filter_attr_1_make=Land%20Rover&filter_attr_2_model=Range%20Rover%20Sport'
response=requests.get(myurl)
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
results=soup.find_all('div',{'class':'masonry-item'})
```

```python
results
```

```python
len(results)
```

```python
#make
htmlcode='<div itemprop="brand" class="b-advert-attribute__value">Land Rover<!----></div>'
makesoup=BeautifulSoup(htmlcode,'html.parser')
brandiv=makesoup.find("div",itemprop='brand')
make=brandiv.get_text(strip=True)
print(make)
```

```python
#model
html_code='<div itemprop="model" class="b-advert-attribute__value">Range Rover Sport<!----></div>'
modelsoup=BeautifulSoup(html_code,'html.parser')
modeldiv=modelsoup.find("div", itemprop='model')
model=modeldiv.get_text(strip=True)
print(model)
```

```python
#manufacture date
datehtml='<div itemprop="productionDate" class="b-advert-attribute__value">2015<!----></div>'
datesoup=BeautifulSoup(datehtml, 'html.parser')
datediv=datesoup.find("div", itemprop='productionDate')
date=datediv.get_text(strip=True)
print(date)
```

```python
#CC
cchtml='<div class="b-advert-attribute__value">3000<span class="h-ml-5">cc</span></div>'
ccsoup=BeautifulSoup(cchtml, 'html.parser')
ccdiv=ccsoup.find("div",{'class':'b-advert-attribute__value'})
cc=ccdiv.get_text(strip=True)
print(cc)
```
